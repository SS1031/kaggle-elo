```python
def recent_5_transactions(x):
    x = x.sort_values('purchase_date', ascending=False)[:5]
    return {
        'recent5_authorize_flag-mean': x.authorized_flag.mean(),
        'recent5_city_id-nunique': x.city_id.nunique(),
        'recent5_merchant_category_id-nunique': x.merchant_category_id.nunique(),
        'recent5_purchase_amount-sum': x.purchase_amount.sum(),
        'recent5_purchase_amount-mean': x.purchase_amount.mean(),
        'recent5_installments-sum': x.installments.sum(),
        'recent5_diff_fam-purchase_date-sum': x['diff_fam-purchase_date'].sum(),
        'recent5_diff_fam-purchase_date-min': x['diff_fam-purchase_date'].min(),
        'recent5_diff_fam-purchase_date-max': x['diff_fam-purchase_date'].max(),
        'recent5_diff_fam-purchase_date-mean': x['diff_fam-purchase_date'].mean(),
        'recent5_diff_refdate-purchase_date-sum': x['diff_refdate-purchase_date'].sum(),
        'recent5_diff_refdate-purchase_date-min': x['diff_refdate-purchase_date'].min(),
        'recent5_diff_refdate-purchase_date-max': x['diff_refdate-purchase_date'].max(),
        'recent5_diff_refdate-purchase_date-mean': x['diff_refdate-purchase_date'].mean(),
    }
```
